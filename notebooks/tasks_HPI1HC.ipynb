**Task** Visualize the p_transient time series for a few more rows.plot

```python
#-----------------------------------------------------
def plot_p_transient(dataTrain,trasients,start,stop):
    plt.figure(figsize=(15,5))
    plt.xticks(rotation=90)

    for i in range(start,stop,1):
        y= dataTrain.loc[i,transients].to_numpy()
        plt.plot(transients,y)

    plt.show()
#---------------------------------------------    
plot_p_transient(dataTrain,transients,1,10)
```

**Domain Knowledge** For our target, the variations within individual time series are more important than across time series.
You come up with the following preprocessing function to adjust your data accordingly:

Quan trong nht vi d liu khng thy ngay t u nu khng co kin thc v cac thng s trn thi khng bit mi quan h nay

```python
def preprocess(df):
    transients = [f"p_transient_{i} [bar]" for i in range(1, 21)]
    # shift the time series, so that the maximum of each series is at 0.
    df.loc[:,transients] = df.loc[:, transients].sub(df.loc[:, transients].max(axis=1), axis=0)
    
    df.loc[:,'minMax'] = df.loc[:, transients].min(axis=1).sub(df.loc[:, transients].max(axis=1), axis=0)

    return df
```

**Task** Apply that function to dataTrain, so that dataTrain contains the preprocessed data.

```python
dataTrain_preprocessed=preprocess(dataTrain)
dataTrain_preprocessed.head()
```

# First ML

**Task** Create a dataframe *dataTrainX* containing only the feature columns.

```python
# auto select the columns by it correlate values
corr_threshold = 0.963 # threshold for corr values, columns has corr higher than corr_threshold will be selected

features=dataTrain_preprocessed.loc[:,transients] # all pressure features

highest_corr_series = corr[abs(features.corrwith(target))>corr_threshold] # select high corr series

high_corr_arr=highest_corr_series.index.to_numpy() # convert to array

high_corr_arr
```

```python
dataTrainX = dataTrain_preprocessed.loc[:,high_corr_arr]

# dataTrainX = dataTrain_preprocessed.iloc[:,1:]

dataTrainX.head()
```

**Task** Create a pandas series *dataTrainY* consisting of only the target column.

```python
dataTrainY=dataTrain_preprocessed.iloc[:,0]
dataTrainY.head()
```

**Task** Import from scikit learn: MinMax-Scaler and (MLP) Multi-Layer Perceptron.

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html

https://scikit-learn.org/stable/modules/neural_networks_supervised.html?highlight=mlp%20multi%20layer%20perceptron

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPRegressor # phan biet regressor and classification
```

**Task** Fit and apply the MinMax-Scaler to *dataTrainX*

```python
scaler = MinMaxScaler()

scaler.fit(dataTrainX)
    
dataTrainX_scale = scaler.transform(dataTrainX)

dataTrainX_scale
```

**Task** Initialize a MLP.

```python
hidden_layer_sizes=(20,10,5)

MLP_R = MLPRegressor(hidden_layer_sizes=hidden_layer_sizes, 
                     max_iter=1000,
                     learning_rate='adaptive', 
                     learning_rate_init=0.1,
                     random_state=2)
```

**Task** Fit the MLP on the training data.

```python
MLP_R.fit(dataTrainX_scale, dataTrainY)
```

```python
MLP_R.score(dataTrainX_scale, dataTrainY)
```
